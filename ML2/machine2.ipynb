### Importing all Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
df=pd.read_csv(r"E:\Datascience\project\ML2\sales_data_with_revenue.csv")
```

### Printing first 20 Rows

```python
print(df.head(20))
```

### Data Cleaning

```python
df=pd.read_csv(r"E:\Datascience\project\ML2\sales_data_with_revenue.csv")
df.columns = [col.replace(" ", "_") for col in df.columns]
df.dropna(inplace=True)
numeric_column = 'numeric_column'
if numeric_column in df.columns:
    df[numeric_column].fillna(df[numeric_column].mean(), inplace=True)
categorical_column = 'categorical_column'
if categorical_column in df.columns:
    df[categorical_column].fillna('Unknown', inplace=True)
df.drop_duplicates(inplace=True)
print(df.shape)
df.isna().sum()
df.info()
```

### Searching for Duplicates

```python
print(df.duplicated().sum())
```

### Statistical Analysis

```python
numeric_data = df.select_dtypes(include=['float64', 'int64'])
stat_an=numeric_data.describe()
print(stat_an)
print(df.head(20))
print(df.tail())
plt.figure(figsize=(15, 10))
sns.boxplot(data=numeric_data)
plt.title('Boxplot of Numeric Data')
plt.xticks(rotation=90)  
plt.show()
```

### Correlation

```python
corelation=numeric_data.corr()
print(corelation)
plt.figure(figsize=(12,8))
sns.heatmap(corelation,cmap='spring',annot=True,fmt='.2f',linewidths=0.5)
plt.title("corelation")
plt.show()
```

### Model Building and Prediction

```python
x=df[['units_sold']]
y=df[['total_revenue']]


plt.scatter(x, y, color="blue", label="Data Points")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter Plot of Data")
plt.legend()
plt.show()

degree = 3  # Degree of the polynomial
poly = PolynomialFeatures(degree=degree)
x_poly = poly.fit_transform(x)


model = LinearRegression()
model.fit(x_poly, y)

y_pred = model.predict(x_poly)

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)*100

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R-squared (R2): {r2:.4f}")


x_new_poly = poly.transform([[6.5]])
y1_pred = model.predict(x_new_poly)
print("The predicted value is:",y1_pred)
plt.scatter(x, y, color="blue", label="Original Data")
plt.plot(x, y_pred, color="red", label=f"Polynomial Degree {degree}")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```

# Summary
### Polynomial regression is an extension of linear regression that models the relationship between independent and dependent variables using polynomial equations. By transforming the original feature x into polynomial features of a given degree, the model captures nonlinear patterns in the data. This approach is particularly useful when data points follow a curved trend that cannot be accurately represented by a simple straight line.  

### In the given implementation, a dataset is loaded, and polynomial regression is applied to model the relationship between two variables. The data is first transformed using polynomial features, then fitted with Linear Regression. The model's performance is evaluated using Mean Squared Error (MSE) and R-squared (R2) scores.  

### Finally, the model is used to make predictions for new data points and visualize the results. A scatter plot displays the original data, while the fitted curve represents the polynomial regression model. The code ensures proper data reshaping and avoids common errors, providing an effective way to analyze and predict nonlinear trends.
